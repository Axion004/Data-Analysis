# Project: Storytelling Data Visualization on Exchange Rates

# Introducing the Dataset

The dataset we'll use describes Euro daily exchange rates between 1999 and 2021. The euro (symbolized with €) is the official currency in most of the countries of the European Union. Daria Chemkaeva put together the data set and made it available on [Kaggle](https://www.kaggle.com/lsind18/euro-exchange-daily-rates-19992020) — the data source is the European Central Bank. 

We will start by reading the dataset.

```python
import pandas as pd

exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')
exchange_rates.head()
```

```python
exchange_rates.tail()
```

```python
exchange_rates.info()
```

Through inspecting the dataset, we can see that there are 5699 rows and 41 columns. All of the columns contain valid (non-null) data and are composed of the object (string) and float data types. The 41 columns correspond to the exchange rate of currency for forty one different countries.

# Data Cleaning

Before we start data visualization, we'll need to do a bit of data cleaning. We will focus on the exchange rate between the euro and the American dollar. More specifically:

- We rename the `[US dollar]` and `Period\Unit:` columns to something easier to type — `US_dollar` and `Time`.
- We change the `Time` column to a `datetime` data type.
- We sort the values by `Time` in ascending order.
- We reset the index (and drop the initial index).

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
```

```python
# Isolate the Time and US_dollar columns
euro_to_dollar = exchange_rates[['Time','US_dollar']]

# Analyze the US_Dollar column to see if the data is stored incorrectly
euro_to_dollar['US_dollar'].value_counts()
```

We now will remove all the `-` characters in the `US_dollar` column and convert the data into floats.

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
euro_to_dollar.info()
```

# Rolling Mean

Now that we're finished cleaning the data, we'll generate a line plot to visualize the evolution of the euro-dollar exchange rate.

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['US_dollar'])
plt.show()
```

If we look at the line's shape, we see many small wiggles — rather than seeing a smooth line. The wiggles, however, have meaning: they are the visual representation of the daily variation in the exchange rate. The rate goes up and down, up and down again, day to day. The rate only shows clear upward or downward trends in the longer run (months or years).

We will now hide the daily variation on our graph and show only the long-term trends using the rolling mean (also known as the moving average).

```python
# Calculate the rolling window from the original data using a window
# of 1 day, 7 days, 30 days, 50 days, 100 days, and 365 days

plt.figure(figsize=(9,6))

plt.subplot(3,2,1)
plt.plot(euro_to_dollar['Time'],euro_to_dollar['US_dollar'])
plt.title("Original Values", weight="bold")

for i, rolling_mean in zip([2,3,4,5,6],
                           [7,30,50,100,365]):
    plt.subplot(3,2,i)
    plt.plot(euro_to_dollar['Time'],
             euro_to_dollar['US_dollar'].rolling(rolling_mean).mean())
    plt.title("Rolling Window: {}".format(rolling_mean), weight="bold")

plt.tight_layout() # Pad the subplots
plt.show()
```

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
euro_to_dollar
```

# Coronavirus Pandemic

First, we will show how the  euro-dollar rate changed during the pandemic using the 2016-2019 data as a baseline.

We will break the dataset into three sections:

- section_1: 2016-2020 (pre-covid)
- section_2: 2020 - 2022 (covid)
- section_3: 2022 - Present (post-covid)

```python
year = euro_to_dollar['Time'].dt.year
pandemic_euro = euro_to_dollar.copy()[(year >= 2016) & (year <= 2021)]
pandemic_euro_2020 = euro_to_dollar.copy()[(year >= 2020) & (year <= 2021)]

# Find the rolling mean
round(pandemic_euro_2020['rolling_mean'].min(),2)
```

```python
# adding the style
import matplotlib.style as style
style.use('fivethirtyeight')

#generating a line plot
fig,ax = plt.subplots(figsize=(10,5))
ax.plot(pandemic_euro['Time'],
         pandemic_euro['rolling_mean'],lw=1, color= 'blue')

# adding the highlight part for pandemic period
ax.plot(pandemic_euro_2020['Time'],
         pandemic_euro_2020['rolling_mean'],lw=2.5, color= 'red')

#changing the color of tick labels
ax.tick_params(axis ='x', colors = 'grey')
ax.tick_params(axis ='y', colors = 'grey')

# removing labels on both axes
ax.set_xticklabels([])
ax.set_yticklabels([])

#to assign new labels we need to find out the tic's coordinates.
#the follwing functions were used:
#ax.get_xticks()
#the output: array([735599., 735964., 736330., 736695., 737060., 737425., 737791.])
#ax.get_yticks()
#the output: array([1.025, 1.05 , 1.075, 1.1  , 1.125, 1.15 , 1.175, 1.2  , 1.225,1.25 ])

#assigning new labes
x = 735964
for year in  ['2016', '2017', '2018', '2019', '2020', '2021']:
    ax.text(x, 1.03, year, alpha=0.5, fontsize=15)
    x += 365
    
y = 1.05
for rate in ['1.05', '1.1', '1.15', '1.2','1.25']:
    ax.text(735800, y, rate, alpha=0.5, fontsize=15)
    y += 0.05

# adding title and labels
ax.text(735800,1.28,'EURO - USD rate plummeted to 1.08 during the COVID-19 pandemic in 2020 ',
        weight = 'bold',size=16)
ax.text(735800, 1.265, 'Euro-USD exchange rates between 2016 and 2021',
        size=14)
ax.text(737418,1.25,'Coronavirus Pandemic',
        weight = 'bold', size=15, rotation = 0)
ax.text(735685,1.15,'Rate',color ='black',alpha=0.3, size=15, rotation = 90)
ax.text(736850,1.01,'Year',color ='black',alpha=0.3, size=15, rotation = 0)

# highlighting the period of pandemic
ax.axvspan(xmin=737425, xmax=737791, ymin=0.03,
           alpha=0.3, color='grey')

# adding the data source
ax.text(735650, 0.99, '©DATAQUEST' + ' '*90 + 'Source: European Central Bank',
        color = '#f0f0f0', backgroundcolor = '#4d4d4d',
        size=15)
plt.show()
```
